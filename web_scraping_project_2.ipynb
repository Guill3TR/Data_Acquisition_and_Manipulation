### **SCRAPING TO THE WEB QUOTES TO SCRAPE**

```python
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://quotes.toscrape.com/")

soup = BeautifulSoup(result.content, 'html.parser')

print(soup.prettify())
```

```python
container = soup.find('div') #finding where all the quotes are
container
```

```python
quotes = container.find_all('span', itemprop="text") #select the 10 quotations on the first page
quotes
```

#### **USING SELENIUM TO ACQUIRE ALL QUOTES FROM THE WEBSITE** ####

```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

options = Options()

options.add_argument('--incognito')

base_url = "https://quotes.toscrape.com/"

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
driver.get(base_url)
```

```python
all_quotes = []

page_index = 1

while True:
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    quotes = soup.find_all('span', itemprop="text")
    all_quotes.extend([quote.get_text() for quote in quotes])

    try:

        next_page_link = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, f"//li[@class='next']/a"))
        )
        next_page_link.click()

        # wait for a specific item to appear on the page after clicking on it
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, 'text')))
        
        page_index += 1
    
    except:

        break


for i, e in enumerate(all_quotes, start=1): #enumerate contains the iterable and the corresponding element
    print(f"{i}. {e}")


driver.quit()
```
